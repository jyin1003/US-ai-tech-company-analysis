```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# interest rate, options, inverse ETF
data = {
    'Data Type': [0.3, 0.5, 0.7], 
    'Simplicty': [0.6, 0.4, 0.3]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create a scatterplot with Seaborn
plt.figure(figsize=(8, 6))

# Set style
sns.set_theme(style="white")

# Add horizontal and vertical lines at the midpoint
plt.axhline(y=0.5, color='black', linestyle='-', linewidth=2)
plt.axvline(x=0.5, color='black', linestyle='-', linewidth=2)


# Set axis limits
plt.xlim(0, 1)
plt.ylim(0, 1)

# Add labels to quadrants
plt.text(0.35, 1.03, 'Big Data Analytics (Lakes)', fontsize=12, color='black')
plt.text(0.3, -0.04, 'Data Warehousing (Warehouses)', fontsize=12, color='black')
plt.text(0.78, 0.52, 'Complex Configurations', fontsize=12, color='black')
plt.text(-0.03, 0.52, 'Simple Setup', fontsize=12, color='black')

# Create a color bar with gradient from white to blue
cmap = plt.get_cmap('Blues')
norm = plt.Normalize(vmin=0, vmax=1)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])

# hide outer axes
sns.despine() 
plt.axis('off')
# Show plot
plt.show()
```
