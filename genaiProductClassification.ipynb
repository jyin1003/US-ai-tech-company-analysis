```python
# Calculate percentages
total_companies = len(companies)
percentages = {cat: len(items) / total_companies * 100 for cat, items in categories.items()}
percentages = {'LLM': 13.88888888888889, 'Product-wide Integration': 44.44444444444444, 'Infrastructure Support Services': 25.0, 'Copilot/Assistant/Chatbots': 38.88888888888889}

# Plotting
fig, ax = plt.subplots(figsize=(9, 10))

# Colors for each category
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']

# Position circles
x = [-0.3, 0.5, 0.8, 1.1]
y = [0.5] * 4

for i, (category, percentage) in enumerate(percentages.items()):
    radius = percentage / 100 
    circle = plt.Circle((x[i], y[i]), radius, color=colors[i], alpha=0.6)
    ax.add_artist(circle)
    
    # Add category label and percentage
    ax.text(x[i], y[i] + radius + 0.05, f"{category} ({percentage:.1f}%)", ha='center', va='bottom', fontweight='bold')
    
    # Add items in the category
    items = categories[category]
    item_text = '\n'.join(items)
    fontsize = 6 if i == 0 else 11 if i == 1 else 10 if i == 3 else 8
    ax.text(x[i], y[i], item_text, ha='center', va='center', fontsize=fontsize, wrap=True)

ax.set_xlim(-0.5, 1.5)
ax.set_ylim(-0.5, 1.5)
ax.set_aspect('equal')  # Ensure the aspect ratio is equal
ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Plotting
fig, ax = plt.subplots(figsize=(34, 6))

# Colors for each category
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']

# Position circles horizontally
x = [0.1, 0.42, 0.77, 1.1]
y = [0.5] * 4

for i, (category, percentage) in enumerate(percentages.items()):
    radius = percentage / 100 / 2
    circle = plt.Circle((x[i], y[i]), radius, color=colors[i], alpha=0.6)
    ax.add_artist(circle)
    
    # Add category label and percentage
    if i == 2:
        ax.text(x[i], y[i] + radius + 0.1, f"{category} ({percentage:.1f}%)", ha='center', va='bottom', fontweight='bold')
    elif i % 2 == 0:
        ax.text(x[i], y[i] + radius + 0.05, f"{category} ({percentage:.1f}%)", ha='center', va='bottom', fontweight='bold')
    else: # stagger labels
        ax.text(x[i], y[i] - radius - 0.05, f"{category} ({percentage:.1f}%)", ha='center', va='top', fontweight='bold')

    # Add items in the category
    items = categories[category]
    item_text = '\n'.join(items)
    fontsize = 6 if i == 0 else 9.5 if i == 1 else 8 if i == 3 else 6
    ax.text(x[i], y[i], item_text, ha='center', va='center', fontsize=fontsize, wrap=True)

ax.set_xlim(0, 1.3)
ax.set_ylim(0, 1)
ax.set_aspect('equal')  # Ensure the aspect ratio is equal
ax.axis('off')

plt.tight_layout()
plt.show()
```
