```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# from:
# https://www.bloomberg.com/company/press/generative-ai-to-become-a-1-3-trillion-market-by-2032-research-finds/

# data
years_software = np.array([2022, 2024, 2026, 2028, 2030, 2032])
software_spending = np.array([1, 5, 11, 22, 37, 62, 97, 149, 204, 261, 318])
software_percent = np.array([0, 1, 1, 2, 3, 4, 6, 8, 9, 11, 11])
years_spending = np.array([2020, 2022, 2024, 2026, 2028, 2030, 2032])
spending_revenue = np.array([11, 18, 37, 64, 128, 219, 340, 457, 611, 790, 981, 1166, 1361])
spending_percent = np.array([0, 2, 4, 6, 7, 8, 10, 11, 12, 12, 12, 12, 12])

fig, ax1 = plt.subplots(figsize=(12, 8))

# plotting the Generative AI Software Spending
ax1.bar(years_software, software_spending[:len(years_software)], color='white', label='Generative AI Software Spending')
ax1.set_xlabel('Year')
ax1.set_ylabel('Revenue ($B)', color='white')
ax1.tick_params(axis='y', labelcolor='white')
ax1.tick_params(axis='x', labelcolor='white')
ax1.set_facecolor('black')
ax1.plot(years_software, software_percent[:len(years_software)], color='orange', label='Gen. AI Software as a % of Total Software')

# second y-axis
ax2 = ax1.twinx()
ax2.set_ylabel('% of Total Software', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')
ax2.set_facecolor('black')

# plotting the Generative AI Revenue
ax1.bar(years_spending - 0.5, spending_revenue[:len(years_spending)], color='blue', width=0.5, label='Generative AI Revenue')
ax2.plot(years_spending, spending_percent[:len(years_spending)], color='red', label='Generative AI as a % of Total Technology Spend')

fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))
plt.title('Generative AI Software Spending and Generative AI Revenue', color='white')
fig.patch.set_facecolor('black')
plt.figtext(0.5, 0.01, 'Source: Bloomberg Intelligence, IDC, eMarketer, Statista', ha='center', color='white')

plt.show()
```
