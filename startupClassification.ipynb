```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# CONSTANTS
FIVE_MILLION = 5000000
TWENTY_MILLION = 20000000
FIFTY_MILLION = 50000000
HUNDRED_MILLION = 100000000
FIVE_HUNDRED_MILLION = 500000000
ONE_BILLION = 1000000000

data = pd.read_excel("Startups Revenue.xlsx")
print(data)
```

```python
# create revenue buckets
less_than_5M = []
from_5M_to_20M = []
from_20M_to_50M = []
from_50M_to_100M = []
from_100M_to_500M = []
above_1B = []

# sort startups into buckets
for index, row in data.iterrows():
    if row["Revenue PA"] <= FIVE_MILLION:
        less_than_5M.append(row["Companies"])
    elif row["Revenue PA"] <= TWENTY_MILLION:
        from_5M_to_20M.append(row["Companies"])
    elif row["Revenue PA"] <= FIFTY_MILLION:
        from_20M_to_50M.append(row["Companies"])
    elif row["Revenue PA"] <= HUNDRED_MILLION:
        from_50M_to_100M.append(row["Companies"])
    elif row["Revenue PA"] <= FIVE_HUNDRED_MILLION:
        from_100M_to_500M.append(row["Companies"])
    else:
        above_1B.append(row["Companies"])

print(less_than_5M)
print(from_5M_to_20M)
print(from_20M_to_50M)
print(from_50M_to_100M)
print(from_100M_to_500M)
print(above_1B)
```

```python
import textwrap

# constant value to create pyramid shape
PYRAMID_SHAPE = [27, 22, 17, 12, 7, 4]

# data for each level
# data = ['less_than_5M', 'from_5M_to_20M', 'from_20M_to_50M', 'from_50M_to_100M', 'from_100M_to_500M', 'above_1B']
data = ['<5M', '5M-20M', '20M-50M', '50M-100M', '100M-500M', '>1B']

# colour palette for the pyramid 
colors = sns.color_palette("Greys", len(data))

# plot the bars 
fig, ax = plt.subplots(figsize=(12, 7))
for i, (category, value, color) in enumerate(zip(data, PYRAMID_SHAPE, colors)):
    ax.barh(category, value, color=color)
    ax.barh(category, -value, color=color)

# remove spines to just show pyramid
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.grid(False)

# add data to each level
for i, label in enumerate(data):
    text_width = PYRAMID_SHAPE[i] * 4
    wrapped_text = textwrap.fill(pyramid_data_company_only[label], width=text_width)
    if (i == 5 or i == 4 or i == 3):
        wrapped_text = textwrap.fill(pyramid_data_company_only[label], width=text_width)
        ax.text(0, i, wrapped_text, va='center', ha='center', color='white', fontsize=9, wrap=True)
    else:
        ax.text(0, i, wrapped_text, va='center', ha='center', color='black', fontsize=10, wrap=True)

# add labels on the left side of the y-axis
ax.set_yticks(range(len(data)))
ax.set_yticklabels(data)
ax.tick_params(axis='y', which='major', pad=20, length=0)  # Increase pad to move labels further left

# configure layout
ax.xaxis.set_ticks([])
plt.tight_layout()
plt.show()
```
